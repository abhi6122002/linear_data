```python
import pandas as pd
from sklearn.linear_model import LinearRegression   
from sklearn.model_selection import  train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\abhishek s\Desktop\machine_learning\Salary_dataset.csv",index_col=0)
data
```

```python
dd=pd.DataFrame(data)
dd
```

```python
dd['Salary'].astype({'Salary': int})
```

```python
dd.dropna()
```

```python
#plt.plot(dd['YearsExperience'],dd['Salary'])
plt.scatter(dd['YearsExperience'],dd['Salary'])
```

```python
x=dd[['YearsExperience']]
y=dd['Salary']
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
ypred=model.predict(xtest)
ypred
```

```python
plt.scatter(xtest,ytest)
plt.scatter(xtest,ypred)
```

```python
r2_score(ytest,ypred)
```

```python
model.score(xtest,ytest)
```

```python
mean_squared_error(ytest,ypred)
```

```python
mean_absolute_error(xtest,ytest)
```
